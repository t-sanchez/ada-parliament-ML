```python
import numpy as np
import pandas as pd
import json
```

```python
df = pd.read_json('links.json')
with open("GroupList.json") as tweetfile:
    nodes = json.load(tweetfile)
```

```python
df.head()
```

```python
jsons = {}
jsons['nodes'] = []
jsons['links'] = []
max_val = 0.23
#for index,row in df.iterrows() :
    


for key in nodes :
    jsons['nodes']+=[{'id':key,'group':nodes[key]}];
    ser_tmp =  df[(df['source']==key) & (df['value']>0)].value.nsmallest(3)
    for entry in ser_tmp.iteritems() :
        jsons['links']+=[({'source':key, 'target':df.target[entry[0]],'value':entry[1]})]
```

```python
with open('data.json', 'w') as outfile:
    json.dump(jsons, outfile)
```
